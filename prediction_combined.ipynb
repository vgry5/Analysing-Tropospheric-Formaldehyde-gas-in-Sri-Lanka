```python
import pandas as pd
```

```python
predictions_DF = []

csv_names = ['Final_HCHO_Pred/Colombo_HCHO_Pred.csv', 'Final_HCHO_Pred/Jaffna_HCHO_Pred.csv', 'Final_HCHO_Pred/Kandy_HCHO_Pred.csv', 'Final_HCHO_Pred/Kurunegala_HCHO_Pred.csv', 'Final_HCHO_Pred/Matara_HCHO_Pred.csv', 'Final_HCHO_Pred/Monaragala_HCHO_Pred.csv', 'Final_HCHO_Pred/Nuwara_E_HCHO_Pred.csv']

for file in csv_names:
    data_frame = pd.read_csv(file)
    predictions_DF.append(data_frame)

predi_combined = pd.concat(predictions_DF, ignore_index=True)
predi_combined.reset_index(drop=True, inplace=True)

predi_combined.to_csv('pred_combined_cleaned.csv', index=False)
predi_combined
```
